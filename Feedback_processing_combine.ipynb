```python
from docx import Document
import pandas as pd
import re

# Load the .docx file
doc_path = "/content/drive/MyDrive/AI Mock Interviewer/DATA/feedback generation.docx"  # Replace with actual path
doc = Document(doc_path)

# Store extracted results
results = []

# Regex patterns
feedback_pattern = r"Feedback:\s*(.*?)(?:, Follow-up:|$)"
followup_pattern = r"Follow-up:\s*(.*)"

# Extract lines from docx
for para in doc.paragraphs:
    text = para.text.strip()
    if not text:
        continue  # skip empty lines

    feedback_match = re.search(feedback_pattern, text)
    followup_match = re.search(followup_pattern, text)

    feedback = feedback_match.group(1).strip() if feedback_match else ""
    followup = followup_match.group(1).strip() if followup_match else ""

    results.append({
        "feedback": feedback,
        "follow_up_question": followup
    })

# Create DataFrame
df = pd.DataFrame(results)

# Optionally, save to CSV
df.to_csv("/content/drive/MyDrive/AI Mock Interviewer/DATA/extracted_feedback_followup.csv", index=False)

print("Extraction complete. Sample rows:")
print(df.head())
```

```python
# Remove rows where all elements are NaN
df = df.dropna(how='all')

# Optional: Remove rows where all fields are empty strings or whitespace
df = df[~(df.applymap(lambda x: isinstance(x, str) and x.strip() == '').all(axis=1))]
```

```python
# Select rows 392 to 1158 (inclusive) from df1
df1_selected = df1.iloc[392:1159].reset_index(drop=True)

# Select all rows from df
df_selected = df.reset_index(drop=True)

# Ensure the number of rows match
min_len = min(len(df1_selected), len(df_selected))
df1_selected = df1_selected.iloc[:min_len]
df_selected = df_selected.iloc[:min_len]

# Concatenate side-by-side (column-wise)
df3 = pd.concat([df1_selected, df_selected], axis=1)
```
